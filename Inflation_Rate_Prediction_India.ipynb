### 6. ip/op

```python
ip = df.drop('Inflation_Rate',axis=1)
op = df.Inflation_Rate
```

```python
ip
```

```python
op
```

### 7. train_test_split

```python
# train_test_split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(ip,op,test_size=0.2)
```

### 8. Standard Scaler

```python
# StandardScaler
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

# 9. MODEL IMPLEMENTATION

## 9.1. Linear Regression

```python
# Model
from sklearn.linear_model import LinearRegression
lr = LinearRegression(n_jobs=-1)
lr.fit(x_train,y_train)
```

```python
# prediction
lr_pred = lr.predict(x_test)
lr_pred
```

```python
# metrics
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
r2_lr = r2_score(y_test,lr_pred)
rmse_lr = np.sqrt(mean_squared_error(y_test,lr_pred))
mae_lr = mean_absolute_error(y_test,lr_pred)
print('r2 score:',r2_lr)
print('rmse score:',rmse_lr)
print('mae score:',mae_lr)
```

```python
# dataframe
dframe = pd.DataFrame({'Actual_Data':y_test,'Predicted':lr_pred})
dframe.head()
```

```python
# plot
sns.lmplot(dframe,x='Actual_Data',y='Predicted',scatter_kws={'color':'blue'},line_kws={'color':'red'})
plt.show()
```

### Future Prediction

```python
ip.tail(3)
```

```python
op.tail(3)
```

```python
# Future Prediction
future = pd.DataFrame({
    'Year':[2026],
    'Month':[1],
    'Brent_Oil_Price_(US/b)':[63],
    'USD_INR':[90],
    'Repo_Rate':[5.5],
    'Petrol_Price':[107],
    'Rainfall':[2]
})
f_scale = sc.transform(future)
prediction = lr.predict(f_scale)
print('Predicted inflation rate of India:',prediction[0])
```

```python
# residual test
residuals_lr = y_test - lr_pred

plt.scatter(lr_pred,residuals_lr)
plt.axhline(y=0, color='red')
plt.xlabel('Predicted Inflation Rate')
plt.ylabel('Residuals')
plt.title('Residual Plot')
plt.show()
```

```python
# coefficient data
coef = pd.DataFrame({
    'Features':ip.columns,
    'Coefficient':lr.coef_
})
print(coef)
```

# Conclusion

```python
results = pd.DataFrame({
    'Model':['Linear','Ridge','KNN','Random Forest'],
    'R2 score':[0.174,0.138,0.568,0.69],
    'RMSE':[1.562,1.595,1.129,0.948],
    'MAE':[1.290,1.332,0.835,0.736]
})
print(results)
```

This project focused on predicting India’s inflation rate using real-world macroeconomic indicators such as Brent oil price, USD/INR exchange rate, repo rate, petrol price, rainfall, month, and year. The dataset was manually collected and combined from multiple government and economic sources, making the project more research-oriented and practical.

Several machine learning models were implemented and compared, including Linear Regression, Ridge Regression, KNN Regressor, and Random Forest Regressor. Performance evaluation was done using R² Score, RMSE, and MAE.

Among all models, Random Forest Regressor achieved the best performance with:

1. R² Score ≈ 0.690
2. RMSE ≈ 0.948
3. MAE ≈ 0.736

The results indicate that inflation prediction involves nonlinear relationships, which were captured more effectively by ensemble learning methods compared to linear models.

The project also revealed economically meaningful relationships:

1. Higher oil prices and USD/INR values tend to increase inflation.
2. Higher repo rates help reduce inflation.
3. Rainfall showed a slight inverse effect on inflation due to agricultural impact.

Overall, this project demonstrates how machine learning can be applied to economic forecasting using real-world data and provides a foundation for future improvements such as larger datasets, additional macroeconomic indicators, and deployment as a real-time forecasting system.
